```python
#collapse
import torch
import numpy as np
import matplotlib.pyplot as plt
co = torch.load('D:\\google\\gan_proj\\framework\\zhq\\new_gan_exp\\repo\\rlt\\collapse\\celeba_scores_0.pth')
x = np.arange(1,51,1)
```

```python
emd = co[1][:, 0]
mmd = co[1][:, 1]
knnreal = co[1][:, 2]
knnfake = co[1][:, 3]
```

```python
incep = co[2][:, 0]
mode = co[2][:, 1]
fid = co[2][:, 2]
```

```python
emd = emd/emd[-1]
mmd = mmd/mmd[-1]
incep = 1-incep/incep[0]
mode = 1-mode/mode[0]
fid = fid/fid[-1]
```
